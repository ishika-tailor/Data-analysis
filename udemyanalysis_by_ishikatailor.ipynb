# 5. Data visualization on udemy data

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
sns.set()
sns.relplot(x="num_subscribers",y="level",data=df)
```

```python
sns.catplot(x="hour",y="subject",kind='bar',data=df);
```
