```python
content=[]
for p in para:
    print(p.contents)
    content.append(p.contents)
```

```python
len(para)
```

```python
usd=[]
percen=[]
```

# USD

```python
for p in para:
    x = re.search("[U][S][D]\s[0-9]+[.]*[0-9]+\s[bBm][i][l][l][i][o][n]",p.contents[0])
    print(x.group())
    usd.append(x.group())
```

```python
import numpy as np
usd1=np.asarray(usd)
usd1
```

```python
len(usd1)
```

# Percentage

```python
for p in para:
    y=re.search("[0-9]+[.]*[0-9\s]+[%]",p.contents[0])
    print(y.group())
    percen.append(y.group())
```

```python
def valid_date(dt):
    result=re.search("[A-Z]+[a-z]+\s[0-9]+", dt)
    return result.group()
```

```python
merge['valid_date']=merge['date'].apply(lambda dt: valid_date(dt))
```
